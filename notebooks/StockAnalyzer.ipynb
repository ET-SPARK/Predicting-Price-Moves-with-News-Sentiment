```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from src.stock_analyzer import StockAnalyzer



symbols = ['AAPL', 'AMZN', 'GOOG', 'META', 'MSFT', 'NVDA', 'TSLA']
base_path = '../data/yfinance_data'

for symbol in symbols:
    file_path = os.path.join(base_path, f"{symbol}_historical_data.csv")
    print(f"\n===== {symbol} =====")
    analyzer = StockAnalyzer(symbol, file_path)
    
    if analyzer.load_data():
        analyzer.calculate_indicators()
        analyzer.visualize()
    else:
        print(f"Skipping {symbol} due to data load failure.")

```
